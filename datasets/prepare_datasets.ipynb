```python
import pandas as pd
```

```python
# remove column "Violations" from chi_food_data.csv
chi_df = pd.read_csv('chi_food_data.csv')

if 'Violations' in chi_df.columns:
    new_chi_df = chi_df.drop(columns=['Violations'])

new_chi_df
```

```python
#save new chi_food_inspection
new_chi_df.to_csv('chi_food_inspection_datasets/chi_food_full_data.csv', index=False)
```

```python
# create ppp datasets: chunk into 20 rows - 22 tables
ppp_df = pd.read_csv('ppp_data.csv')
dataframes = [ppp_df.sample(n=20, random_state=i).reset_index(drop=True) for i in range(22)]
# Save each DataFrame in the list to a separate CSV file
for i, df in enumerate(dataframes, start=1):
    filename = f'ppp_datasets/ppp_data_{i}.csv'
    df.to_csv(filename, index=False)
```

```python
# create chicago datasets: chunk into 10 rows - 13 tables
# chi_df = pd.read_csv('chi_food_inspection_datasets/chi_food_full_data.csv')
dataframes = [new_chi_df.sample(n=10, random_state=i).reset_index(drop=True) for i in range(13)]
# Save each DataFrame in the list to a separate CSV file
for i, df in enumerate(dataframes, start=1):
    filename = f'chi_food_inspection_datasets/chi_food_data_{i}.csv'
    df.to_csv(filename, index=False)
```

```python
dish_df = pd.read_csv('dish_data.csv')
```

```python
dish_df_pre = dish_df.drop(columns=['description'])
```

```python
dish_df_pre
```

```python
# create dish datasets: chunk into 50 rows ~ 16 tables
dataframes = [dish_df_pre.sample(n=50, random_state=i).reset_index(drop=True) for i in range(16)]
# Save each DataFrame in the list to a separate CSV file
for i, df in enumerate(dataframes, start=1):
    filename = f'dish_datasets/dish_data_{i}.csv'
    df.to_csv(filename, index=False)
```
